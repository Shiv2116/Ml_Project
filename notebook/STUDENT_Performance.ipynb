## Multivariate Analysis 

```python
# Set the global size of the figure for all plots
plt.rcParams['figure.figsize'] = (30, 12)  # Set figure width to 30 and height to 12

# Subplot 1: Gender distribution
plt.subplot(1, 5, 1)  # Create the first subplot in a 1x5 grid (position 1)
size = df['gender'].value_counts()  # Count the occurrences of each gender in the DataFrame
labels = 'Female', 'Male'  # Define the labels for the pie chart
color = ['red', 'green']  # Define colors for the slices

plt.pie(size, colors=color, labels=labels, autopct='%.2f%%')
# Plot the pie chart
plt.title('Gender', fontsize=20)  # Add a title to the pie chart
plt.axis('off')  # Turn off axis lines for a cleaner look

# Subplot 2: Race/Ethnicity distribution
plt.subplot(1, 5, 2)  # Create the second subplot in the grid (position 2)
size = df['race_ethnicity'].value_counts()  # Count occurrences of each race/ethnicity group
labels = 'Group C', 'Group D', 'Group B', 'Group E', 'Group A'  # Labels for the groups
color = ['red', 'green', 'blue', 'cyan', 'orange']  # Define colors for the slices

plt.pie(size, colors=color, labels=labels, autopct='%.2f%%')# Plot the pie chart
plt.title('Race_Ethnicity', fontsize=20)  # Add a title
plt.axis('off')  # Turn off axis lines

# Subplot 3: Lunch distribution
plt.subplot(1, 5, 3)  # Create the third subplot in the grid (position 3)
size = df['lunch'].value_counts()  # Count occurrences of each lunch category
labels = 'Standard', 'Free'  # Labels for lunch types
color = ['red', 'green']  # Define colors for the slices

plt.pie(size, colors=color, labels=labels, autopct='%.2f%%')# Plot the pie chart
plt.title('Lunch', fontsize=20)  # Add a title
plt.axis('off')  # Turn off axis lines

# Subplot 4: Test Preparation Course distribution
plt.subplot(1, 5, 4)  # Create the fourth subplot in the grid (position 4)
size = df['test_preparation_course'].value_counts()  # Count occurrences of test course status
labels = 'None', 'Completed'  # Labels for the test preparation course
color = ['red', 'green']  # Define colors for the slices
plt.pie(size, colors=color, labels=labels, autopct='%.2f%%')
 # Plot the pie chart
plt.title('Test_Course', fontsize=20)  # Add a title
plt.axis('off')  # Turn off axis lines

# Subplot 5: Parental Level of Education distribution
plt.subplot(1, 5, 5)  # Create the fifth subplot in the grid (position 5)
size = df['parental_level_of_education'].value_counts()  # Count occurrences of each education level
labels = 'Some College', "Associate's Degree", 'High School', 'Some High School', \
         "Bachelor's Degree", "Master's Degree"  # Labels for education levels
color = ['red', 'green', 'blue', 'cyan', 'orange', 'grey']  # Define colors for the slices

plt.pie(size, colors=color, labels=labels, autopct='%.2f%%')
 # Plot the pie chart
plt.title('Parental_Education', fontsize=20)  # Add a title
plt.axis('off')  # Turn off axis lines

# Adjust layout to prevent overlap between subplots
plt.tight_layout()

# Add gridlines (though gridlines won't affect pie charts)
plt.grid()

# Display the figure
plt.show()
```

## Bivariate Analysis - does race_ethnicity has effect on student performance ?


```python
# Group the data by 'race_ethnicity'
Group_data2 = df.groupby('race_ethnicity')

# Create subplots
f, ax = plt.subplots(1, 3, figsize=(20, 8))

# Plot for Math score
sns.barplot(x=Group_data2['math_score'].mean().index, y=Group_data2['math_score'].mean().values, palette='mako', ax=ax[0])
ax[0].set_title('Math Score', color='#005ce6', size=20)
ax[0].set_xlabel('Race/Ethnicity', size=15)
ax[0].set_ylabel('Average Score', size=15)
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=15)

# Plot for Reading score
sns.barplot(x=Group_data2['reading_score'].mean().index, y=Group_data2['reading_score'].mean().values, palette='flare', ax=ax[1])
ax[1].set_title('Reading Score', color='#005ce6', size=20)
ax[1].set_xlabel('Race/Ethnicity', size=15)
ax[1].set_ylabel('Average Score', size=15)
for container in ax[1].containers:
    ax[1].bar_label(container, color='black', size=15)

# Plot for Writing score
sns.barplot(x=Group_data2['writing_score'].mean().index, y=Group_data2['writing_score'].mean().values, palette='coolwarm', ax=ax[2])
ax[2].set_title('Writing Score', color='#005ce6', size=20)
ax[2].set_xlabel('Race/Ethnicity', size=15)
ax[2].set_ylabel('Average Score', size=15)
for container in ax[2].containers:
    ax[2].bar_label(container, color='black', size=15)

# Adjust layout
plt.tight_layout()
plt.show()
```
